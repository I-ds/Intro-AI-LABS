```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('/content/gdrive/MyDrive/Module_IA_UM6P_LAHOUSSINE_BOUZZIT/LAB_03/exemple_dataset_industriel.csv')
```

```python
data.head()
```

```python
x = data[['Feature_1','Feature_2','Feature_3','Feature_4']]
y = data[['KPI_1','KPI_2']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

lr= LinearRegression()
lr.fit(x_train,y_train)
print("coef:", lr.coef_)
print("intercept:", lr.intercept_ )
```

```python
predict=lr.predict(x_test)
print('MSE : ', metrics.mean_squared_error(y_test,predict))
print('RMSE : ',np.sqrt(metrics.mean_squared_error(y_test,predict)))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(criterion='squared_error')
rf.fit(x_train, y_train)
```

```python
predictions=rf.predict(x_test)
print("mse", metrics.mean_squared_error(y_test, predictions))
print ("RMSE", np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor(criterion ='squared_error', random_state=0)
dt.fit(x_train, y_train)
print(dt.tree_.max_depth)
```

```python
dt.fit(x_train, y_train)
predictions=dt.predict(x_test)
print("MSE", metrics.mean_squared_error(y_test, predictions))
print("RMSE", np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

problème de régression avec deep learning

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = Sequential([
    Dense(units=64, input_dim=x_train.shape[1], activation='relu'),
    Dense(units=2) ])
```

```python
# Compilez le modèle
model.compile(optimizer='adam', loss='mean_squared_error')

# Entraînez le modèle
model.fit(x_train, y_train, epochs=100, verbose=0)

# Effectuez des prédictions sur l'ensemble de test
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)
```

```python
model = tf.keras.Sequential([
    Dense(units=2, input_dim=x_train.shape[1], activation='linear')
])

# Compilez le modèle
model.compile(optimizer='adam', loss='mean_squared_error')

# Entraînez le modèle
model.fit(x_train, y_train, epochs=100, verbose=0)

# Effectuez des prédictions sur l'ensemble de test
y_pred = model.predict(x_test)

# Calculez les métriques de performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)

```
