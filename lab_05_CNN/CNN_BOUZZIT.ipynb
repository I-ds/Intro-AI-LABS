```python
# On importe les librairies dont on aura besoin
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

image = np.array([[0,128,255,0],
                  [255,0,128,255],
                  [128,255,0,128],
                  [0,128,255,0]], dtype = np.uint8
)

plt.imshow(image,cmap="gray")
plt.title("Exemple d'image en niveaux de gris")
plt.axis('off')
plt.show()
```

```python
from scipy.signal import convolve2d
kernel = np.array([[1,0,-1],
                  [2,0,-2],
                  [1,0,-1]]
)

convolution_result = convolve2d(image,kernel,mode="same", boundary="wrap")
```

```python
plt.figure(figsize=(8,4))
plt.subplot(121)
plt.imshow(image,cmap="gray")
plt.title("Image d'origine")
plt.axis('off')

plt.subplot(122)
plt.imshow(convolution_result, cmap='gray')
plt.title("Resultat de la covolution")
plt.axis('off')
```

```python
import numpy as np
from scipy.ndimage import gaussian_filter, sobel
from PIL import Image

image_path= '/content/gdrive/MyDrive/Module_IA_UM6P_LAHOUSSINE_BOUZZIT/lab_04/images.jpeg'
image  = Image.open(image_path)

#convertir l'image en tableau numpy

image= np.array(image)



#appliquer un filtre gaussian

blurred_image = gaussian_filter(image, sigma =1)

#edge detection

edge_detection = sobel(image)
#afficher les résultats:
plt.figure(figsize=(12,4))

plt.subplot(131)
plt.imshow(image)
plt.title("Original image")
plt.axis('off')

plt.figure(figsize=(12,4))
plt.subplot(132)
plt.imshow(blurred_image)
plt.title("Blurred image")
plt.axis('off')

plt.subplot(133)
plt.imshow(edge_detection)
plt.title("Sobel image")
plt.axis('off')
```
